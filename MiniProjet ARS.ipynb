# Fonction voisins communs

```python
def Voisins_Communs(g,x,y):
    Nx= g.neighbors(x)
    Ny = g.neighbors(y)
    return len(list(set(Nx) & set(Ny)))
Voisins_Communs(g,1,4)
```

# Fonction indice de Salton

```python
import math
def Indice_salton(g,x,y):
    Nx= g.neighbors(x)
    Ny= g.neighbors(y)
    return Voisins_Communs(g,x,y)/(math.sqrt(len(Nx)*len(Ny)))
Indice_salton(g,1,4)
```
